# Transfer learning

## [Xception](https://www.kaggle.com/amadeus1996/monkey-classifier-xception-acc-97/notebook)

```python
img_width, img_height = 256,256
train_data_dir = 'train/'
valid_data_dir = 'validation/'
batch_size = 3
num_class = 15
lr = 1e-4
momentum=0.9
epoch=15
```

```python
train_datagen = ImageDataGenerator(rotation_range=30, 
                                   rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)
train_generator = train_datagen.flow_from_directory(train_data_dir,
                                                    target_size=(img_height, img_width),
                                                    batch_size=batch_size,
                                                    class_mode='categorical')

valid_datagen = ImageDataGenerator(rescale=1./255)
valid_generator = valid_datagen.flow_from_directory(valid_data_dir,
                                                    target_size=(img_height, img_width),
                                                    batch_size=batch_size,
                                                    class_mode='categorical')
```

```python
train_generator.class_indices
```

```python
Xception_base = applications.Xception(weights='imagenet', include_top=False)

x = Xception_base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(512,activation='relu')(x)
x = layers.Dense(256,activation='relu')(x)
x = layers.Dense(128,activation='relu')(x)

predictions = layers.Dense(num_class, activation='softmax')(x)

Xception_transfer = models.Model(inputs=Xception_base.input, outputs=predictions)
```

```python
Xception_transfer.compile(loss = 'categorical_crossentropy',
                           optimizer = optimizers.SGD(lr=lr, momentum=momentum),
                           metrics=['accuracy'])
```

```python
from tensorflow.python.client import device_lib
print(device_lib.list_local_devices())
```

```python
import tensorflow as tf
with tf.device('/device:GPU:0'):
    history = Xception_transfer.fit_generator(train_generator,epochs=epoch,shuffle=True,verbose=1,validation_data=valid_generator)
```

```python
test = os.listdir('testset/')
INPUT_SIZE=256
x_test = np.zeros((len(test), INPUT_SIZE, INPUT_SIZE, 3), dtype='float32')
for i in range(len(test)):
    gg = image.load_img('testset/' + test[i],target_size=(INPUT_SIZE, INPUT_SIZE))
    gg = image.img_to_array(gg) / 255
    x_test[i] = gg
print('Test Images shape: {} size: {:,}'.format(x_test.shape, x_test.size))
```

```python
predss = history.model.predict(x_test)
```

```python
predss2 = np.argmax(predss,axis=1)
predss2
```

```python
inverdict = {k:v for v,k in train_generator.class_indices.items()}
predss3 = [inverdict.get(k) for k in predss2]
```

```python
for i in range(len(test)):
    ggg = cv2.imread('testset/' + test[i])
    print(ggg.shape)
```

```python
dd = pd.DataFrame({'class':[correct_dict[k] for k in predss3], 'id':[i.split('.')[0] for i in test]})
dd = dd[['id','class']]
```

## [Extract Xception bottleneck features](https://www.kaggle.com/gaborfodor/dog-breed-pretrained-keras-models-lb-0-3/notebook)

```python
NUM_CLASSES=15
SEED=111
INPUT_SIZE=256
x_train = np.zeros((len(train_generator.filenames), INPUT_SIZE, INPUT_SIZE, 3), dtype='float32')
for i in range(len(train_generator.filenames)):
    gg = image.load_img('train/' + train_generator.filenames[i],target_size=(INPUT_SIZE, INPUT_SIZE))
    gg = image.img_to_array(gg)
    x = xception.preprocess_input(np.expand_dims(gg.copy(), axis = 0))
    x_train[i] = x
print('Train Images shape: {} size: {:,}'.format(x_train.shape, x_train.size))
```

```python
x_valid = np.zeros((len(valid_generator.filenames), INPUT_SIZE, INPUT_SIZE, 3), dtype='float32')
for i in range(len(valid_generator.filenames)):
    gg = image.load_img('validation/' + valid_generator.filenames[i],target_size=(INPUT_SIZE, INPUT_SIZE))
    gg = image.img_to_array(gg)
    x = xception.preprocess_input(np.expand_dims(gg.copy(), axis = 0))
    x_valid[i] = x
print('Validation Images shape: {} size: {:,}'.format(x_valid.shape, x_valid.size))
```

```python
y_train = pd.get_dummies(train_generator.classes.astype(str))
y_valid = pd.get_dummies(valid_generator.classes.astype(str))

print(x_train.shape, x_valid.shape, y_train.shape, y_valid.shape)
```

```python
x_train, y_train = shuffle(x_train, y_train)
x_valid, y_valid = shuffle(x_valid, y_valid)
```

```python
POOLING = 'avg'
xception_bottleneck = xception.Xception(weights='imagenet', include_top=False, pooling=POOLING)
train_x_bf = xception_bottleneck.predict(x_train, batch_size=32, verbose=1)
valid_x_bf = xception_bottleneck.predict(x_valid, batch_size=32, verbose=1)
print('Xception train bottleneck features shape: {} size: {:,}'.format(train_x_bf.shape, train_x_bf.size))
print('Xception valid bottleneck features shape: {} size: {:,}'.format(valid_x_bf.shape, valid_x_bf.size))
```

```python
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=SEED)
logreg.fit(train_x_bf, (y_train * range(NUM_CLASSES)).sum(axis=1))
valid_probs = logreg.predict_proba(valid_x_bf)
valid_preds = logreg.predict(valid_x_bf)
print('Validation Xception LogLoss {}'.format(log_loss(y_valid, valid_probs)))
print('Validation Xception Accuracy {}'.format(accuracy_score((y_valid * range(NUM_CLASSES)).sum(axis=1), valid_preds)))
```

## [Extract Inception bottleneck features](https://www.kaggle.com/gaborfodor/dog-breed-pretrained-keras-models-lb-0-3/notebook)

```python
INPUT_SIZE=299
x_train = np.zeros((len(train_generator.filenames), INPUT_SIZE, INPUT_SIZE, 3), dtype='float32')
for i in range(len(train_generator.filenames)):
    gg = image.load_img('train/' + train_generator.filenames[i],target_size=(INPUT_SIZE, INPUT_SIZE))
    gg = image.img_to_array(gg)
    x = xception.preprocess_input(np.expand_dims(gg.copy(), axis = 0))
    x_train[i] = x
print('Train Images shape: {} size: {:,}'.format(x_train.shape, x_train.size))
```

```python
x_valid = np.zeros((len(valid_generator.filenames), INPUT_SIZE, INPUT_SIZE, 3), dtype='float32')
for i in range(len(valid_generator.filenames)):
    gg = image.load_img('validation/' + valid_generator.filenames[i],target_size=(INPUT_SIZE, INPUT_SIZE))
    gg = image.img_to_array(gg)
    x = xception.preprocess_input(np.expand_dims(gg.copy(), axis = 0))
    x_valid[i] = x
print('Validation Images shape: {} size: {:,}'.format(x_valid.shape, x_valid.size))
```

```python
y_train = pd.get_dummies(train_generator.classes.astype(str))
y_valid = pd.get_dummies(valid_generator.classes.astype(str))

print(x_train.shape, x_valid.shape, y_train.shape, y_valid.shape)
```

```python
x_train, y_train = shuffle(x_train, y_train)
x_valid, y_valid = shuffle(x_valid, y_valid)
```

```python
POOLING = 'avg'
print((x_train.shape, x_valid.shape, y_train.shape, y_valid.shape))
inception_bottleneck = inception_v3.InceptionV3(weights='imagenet', include_top=False, pooling=POOLING)
train_i_bf = inception_bottleneck.predict(x_train, batch_size=32, verbose=1)
valid_i_bf = inception_bottleneck.predict(x_valid, batch_size=32, verbose=1)
print('InceptionV3 train bottleneck features shape: {} size: {:,}'.format(train_i_bf.shape, train_i_bf.size))
print('InceptionV3 valid bottleneck features shape: {} size: {:,}'.format(valid_i_bf.shape, valid_i_bf.size))
```

```python
SEED=20
NUM_CLASSES=15
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=SEED)
logreg.fit(train_i_bf, (y_train * range(NUM_CLASSES)).sum(axis=1))
valid_probs = logreg.predict_proba(valid_i_bf)
valid_preds = logreg.predict(valid_i_bf)
print('Validation Inception LogLoss {}'.format(log_loss(y_valid, valid_probs)))
print('Validation Inception Accuracy {}'.format(accuracy_score((y_valid * range(NUM_CLASSES)).sum(axis=1), valid_preds)))
```

## [LogReg on all bottleneck features](https://www.kaggle.com/gaborfodor/dog-breed-pretrained-keras-models-lb-0-3/notebook)

```python
X = np.hstack([train_x_bf, train_i_bf])
V = np.hstack([valid_x_bf, valid_i_bf])
print('Full train bottleneck features shape: {} size: {:,}'.format(X.shape, X.size))
print('Full valid bottleneck features shape: {} size: {:,}'.format(V.shape, V.size))
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=SEED)
logreg.fit(X, (y_train * range(NUM_CLASSES)).sum(axis=1))
valid_probs = logreg.predict_proba(V)
valid_preds = logreg.predict(V)
print('Validation Xception + Inception LogLoss {}'.format(log_loss(y_valid, valid_probs)))
print('Validation Xception + Inception Accuracy {}'.format(accuracy_score((y_valid * range(NUM_CLASSES)).sum(axis=1), valid_preds)))
```
